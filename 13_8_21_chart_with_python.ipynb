```python
import pandas as pd
data = pd.read_csv("/content/ADANIPORTS.NS.csv")
import plotly.graph_objects as plt
figure = plt.Figure(data=[plt.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])
figure.update_layout(title = "Adani ports Price Analysis", 
                     xaxis_rangeslider_visible=False)
figure.show()
```
